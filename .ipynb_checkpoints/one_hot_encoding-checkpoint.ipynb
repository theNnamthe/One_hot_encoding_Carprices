```python
#import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_info = pd.read_csv("carprices.csv")
```

```python
#view dataset
car_info
```

## SCATTER PLOTS

```python
#Create scatter plot
#Essence of this is to check for linear relationship

# Create a scatter plot with 'Mileage' on the x-axis and 'Sell Price($)' on the y-axis

plt.figure(figsize =(10,6))
sns.scatterplot(x= 'Mileage',y= 'Sell Price($)', hue = 'Car Model',style='Car Model', data=car_info)

# Add titles and labels
plt.title('Scatter Plot of Sell Price vs Mileage by Car Model')
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.legend(title='Car Model')
plt.show()

# Create a scatter plot with 'Age(yrs)' on the x-axis and 'Sell Price($)' on the y-axis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age(yrs)', y='Sell Price($)', hue='Car Model', style='Car Model', data=car_info)

# Add titles and labels
plt.title('Scatter Plot of Sell Price vs Age by Car Model')
plt.xlabel('Age (years)')
plt.ylabel('Sell Price($)')
plt.legend(title='Car Model')
plt.show()
```

We establish that there are linear relationships between Sell prices of the cars and the ages/mileage of the cars 

## Use pandas to create dummy variables 

This essentially is to change the Car Model column from a Categorical-Nominal type to an ordinal type

```python
#dummy dataframe for Car Model

car_model_dummy = pd.get_dummies(car_info['Car Model']) #columns=['Car Model'])
```

```python
car_model_dummy
```

```python
#Merge both dataframes

Merged_car_info = pd.concat([car_info, car_model_dummy], axis= 1)
Merged_car_info
```

```python
#drop redundant Car model column
Final_car_info = Merged_car_info.drop(['Car Model'], axis = 1)
Final_car_info
```

When you can derive one variable from other variables, this is known as multicollinearity. In this case, if you know the values for Audi A5 and BMW X5, you can easily infer the value for Mercedes Benz C class. For instance if Audi A5 = 0 and  BMW X5 = 0, then Mercedes Benz C class must be 1. Therefore, these variables are multicollinear. This situation can cause issues with linear regression, so one column should be dropped to avoid this problem.

**NOTE: I can choose to drop the one variable myself or allow the sklearn library to handle the dummy variable trap, so even if you don't drop one of the state columns, it will still work. However, it is a good practice to handle the dummy variable trap yourself, in case the library you are using does not manage this for you.**

```python
Final_car_info = Final_car_info.drop(['BMW X5'], axis=1)
Final_car_info
```

```python

#reorder the columns
columns = Final_car_info.columns.tolist() #to convert the DataFrame's columns to a list.
columns = columns[:1] + columns[2:] + [columns[1]] # Reorder columns so Sell Price($) is the last column
columns
```

```python
Final_car_info[columns]
```

```python
#Create X and y variables for the linear regression model

X = Final_car_info[['Mileage', 'Age(yrs)', 'Audi A5', 'Mercedez Benz C class']]

y = Final_car_info['Sell Price($)']
```

```python
#Import linear regression model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Fit the model for training 
model.fit(X,y)
```

#### PREDICTIONS

**Predict price of a mercedez benz that is 4 yr old with mileage 45000**

```python
model.predict([[45000,4,0,1]])
```

**Price of BMW X5 that is 7 yr old with mileage 86000**

```python
model.predict([[86000,7,0,0]])
```

```python
model.score(X,y)
```
